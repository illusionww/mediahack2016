```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import matplotlib
%matplotlib inline
```

```python
subjects = []
for subject_id in range(1, 8 + 1):
    years = {}
    for year in range(1993, 2014 + 1):
        url = 'http://www.rfbr.ru/rffi/ru/project_search?query=&GRANT_ITEMS=1' + \
              '&GRANT_SUBJECT_TYPE=' + str(subject_id) + \
              '&GRANT_YEAR=' + str(year) + \
              '&GRANT_TYPE_ID=1'
        
        soup = BeautifulSoup(requests.get(url).text, 'html.parser')
        li = soup.find('li', {'class': 'pager-in item-4 '})
        if (li is None):
            years[year] = 0
        else:
            years[year] = int(li.find('a', {'class': 'pager-in-link'}).text)
    
    subjects.append(years)
```

```python
subjects_names = (
                    'Математика, механика, информатика',
                    'Физика и астрономия',
                    'Химия и науки о материалах',
                    'Биология и медицинские науки',
                    'Науки о Земле',
                    'Естественнонаучные методы исследований в гуманитарных науках',
                    'Инфокоммуникационные технологии и вычислительные системы',
                    'Фундаментальные основы инженерных наук'
                 )
```

```python
pd.DataFrame(subjects, index=subjects_names)
```

```python
pd.DataFrame(subjects, index=subjects_names).to_csv('grants_fields_by_years_all_approved.csv')
```
