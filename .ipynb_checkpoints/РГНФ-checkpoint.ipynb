```python
def getList(offset):
    url = "http://grant.rfh.ru/sys/a/?colfilter=0&context=_anonymous~&fedcols=1&multi=1&pgoffset=" + str(offset) + "&ro_filter=_main.enrfh_tasks.enf_keywordsw%20contains%20-~_main.enrfh_tasks.enf_last_datew%20!%3D%20~_main.enrfh_tasks.enf_last_decisionw%20%3D%201~&sorting=_main.enrfh_tasks.enf_numberw:1~&table=main.enrfh_tasks&target=show_template&template=ext_filter_result.htm"
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    table = soup.find("table")
    aa = table.findAll('a')
    if len(aa) == 0:
        raise Exception(offset)
    links = []
    for a in aa:
        links.append(a['href'])
    return links
```
