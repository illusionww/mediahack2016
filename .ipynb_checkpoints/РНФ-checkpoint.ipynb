```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("rnf.txt") as f:
    text = f.read()
```

```python
soup = BeautifulSoup(text, 'html.parser')
```

```python
def get_name(num):
    if num.startswith('01'):
        return "Математика, информатика и науки о системах"
    elif num.startswith('02'):
        return "Физика и науки о космосе"
    elif num.startswith('03'):
        return "Химия и науки о материалах"
    elif num.startswith('04'):
        return "Биология и науки о жизни"
    elif num.startswith('05'):
        return "Фундаментальные исследования для медицины"
    elif num.startswith('06'):
        return "Сельскохозяйственные науки"
    elif num.startswith('07'):
        return "Науки о Земле"
    elif num.startswith('08'):
        return "Гуманитарные и социальные науки"
    elif num.startswith('09'):
        return "Инженерные науки"
```

* 01 - Математика, информатика и науки о системах
* 02 - Физика и науки о космосе
* 03 - Химия и науки о материалах
* 04 - Биология и науки о жизни
* 05 - Фундаментальные исследования для медицины
* 06 - Сельскохозяйственные науки
* 07 - Науки о Земле
* 08 - Гуманитарные и социальные науки
* 09 - Инженерные науки


```python
items = []
for tr in soup.findAll("tr"):
    tds = tr.findAll("td")
    item = {
        "№": tds[0].text.strip(),
        "Номер проекта": tds[1].text.strip(),
        "Название": tds[2].span.text.strip(),
        "Руководитель": tds[2].em.text.strip(),
        "Ключевые слова": tds[2].span["title"].strip(),
        "Коды классификатора": get_name(tds[3].findAll('span')[0].text.strip()),
        "Конкурс": tds[4].text.strip(),
        "Год": tds[4].text.strip()[8:12],
        "Организация, регион": tds[5].text.strip(),
    }
    items.append(item)
```

```python
df = pd.DataFrame(items)
df = df[["№", "Номер проекта", "Название", "Руководитель", "Ключевые слова", "Коды классификатора", "Конкурс", "Год", "Организация, регион"]]
df
```
